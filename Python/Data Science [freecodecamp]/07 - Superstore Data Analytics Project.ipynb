```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Sample - Superstore.csv", encoding="ISO-8859-1", index_col=0, parse_dates=[2, 3])
df
```

```python
df.info()
```

```python
df = df.drop_duplicates()
df
```

```python
types_of_customers = df["Segment"].unique()
types_of_customers
```

```python
number_of_customers = df["Segment"].value_counts()
number_of_customers.name =  "Type of Customer"
number_of_customers
```

```python
plt.pie(number_of_customers, labels=number_of_customers.index, autopct="%0.1f%%")
plt.show()
```

```python
sales_per_segment = df.groupby("Segment")["Sales"].sum()
sales_per_segment.index.name = "Types of Customer"
sales_per_segment.name = "Total Sales"
sales_per_segment
```

```python
plt.bar(sales_per_segment.index, sales_per_segment)
plt.show()
```

```python
plt.pie(sales_per_segment, labels=sales_per_segment.index, autopct="%1.1f%%")
plt.show()
```

```python
df
```

```python
customer_order_frequency = df.groupby(["Customer ID", "Customer Name", "Segment"])["Order ID"].count().reset_index().set_index("Customer ID")
customer_order_frequency = customer_order_frequency.rename(columns={"Order ID": "Total Orders"})
customer_order_frequency
```

```python
repeated_customers = customer_order_frequency[customer_order_frequency["Total Orders"] > 1]
repeated_customers
```

```python
repeated_customers_sorted = repeated_customers.sort_values("Total Orders", ascending=False)
repeated_customers_sorted
```

```python
customer_sales = df.groupby(["Customer ID", "Customer Name", "Segment"])["Sales"].sum().reset_index().set_index("Customer ID")
customer_sales = customer_sales.rename(columns={"Sales": "Total Sales"})
customer_sales
```

```python
top_spenders = customer_sales.sort_values("Total Sales", ascending=False)
top_spenders
```

```python
df
```

```python
shipping_model = df["Ship Mode"].value_counts()
shipping_model.index.name = "Type of Shipping"
shipping_model.name = "Total Shipping"
shipping_model
```

```python
plt.pie(shipping_model, labels=shipping_model.index, autopct="%1.1f%%")
plt.show()
```

```python
df
```

```python
state = df["State"].value_counts()
state.name = "Total Customers"
state
```

```python
city = df["City"].value_counts()
city.name = "Total Customers"
city
```

```python
state_sales = df.groupby("State")["Sales"].sum()
top_state_sales = state_sales.sort_values(ascending=False)
top_state_sales.name = "Total Sales"
top_state_sales
```

```python
city_sales = df.groupby("City")["Sales"].sum()
top_city_sales = city_sales.sort_values(ascending=False)
top_state_sales.name = "Total Sales"
top_city_sales
```

```python
state_sales_performance = pd.merge(state, state_sales, on="State")
state_sales_performance
```

```python
city_sales_performance = pd.merge(city, city_sales, on="City")
city_sales_performance
```

```python
products = df["Category"].unique()
products
```

```python
products_subcategory = df["Sub-Category"].unique()
products_subcategory
```

```python
subcategory_count = (df.groupby("Category")["Sub-Category"].nunique())
subcategory_count = subcategory_count.sort_values(ascending=False)
subcategory_count
```

```python
subcategory_count_sales = df.groupby(["Category", "Sub-Category"])["Sales"].sum().reset_index().rename(columns={"Sales": "Total Sales"})
subcategory_count_sales = subcategory_count_sales.sort_values("Total Sales", ascending=False)
subcategory_count_sales
```

```python
category_count_sales = df.groupby("Category")["Sales"].sum()
category_count_sales.name = "Total Sales"
category_count_sales = category_count_sales.sort_values(ascending=False)
category_count_sales
```

```python
plt.pie(category_count_sales, labels=category_count_sales.index, autopct="%1.1f%%")
plt.show()
```

```python
subcategory_count_sales = subcategory_count_sales.sort_values("Total Sales")
plt.barh(subcategory_count_sales["Sub-Category"], subcategory_count_sales["Total Sales"])
plt.xlabel("Sales")
plt.ylabel("Sub-Category")
plt.show()
```

```python
yearly_sales = df.groupby(df["Order Date"].dt.year)["Sales"].sum()
yearly_sales.index.name = "Year"
yearly_sales.name = "Total Sales"
yearly_sales
```

```python
plt.bar(yearly_sales.index, yearly_sales)
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.plot(yearly_sales.index, yearly_sales, marker='o', linestyle="-")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
year_sales = df[df["Order Date"].dt.year == 2017]
quarterly_sales = year_sales.resample("QE", on="Order Date")["Sales"].sum()
quarterly_sales.index.name = "Quarterly"
quarterly_sales.name = "Total Sales"
quarterly_sales
```

```python
plt.plot(quarterly_sales.index, quarterly_sales, marker="o", linestyle="--")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
year_sales = df[df["Order Date"].dt.year == 2017]
monthly_sales = year_sales.resample("ME", on="Order Date")["Sales"].sum()
monthly_sales.index.name = "Monthly"
monthly_sales.name = "Total Sales"
monthly_sales
```

```python
plt.plot(monthly_sales.index, monthly_sales, marker="o", linestyle="--")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```
