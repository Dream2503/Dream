1.  Write a program to input electricity unit charges and calculate total electricity bill according to the given condition:<br>
        For first 50 units Rs. 0.50/unit<br>
        For next 100 units Rs. 0.75/unit<br>
        For next 100 units Rs. 1.20/unit<br>
        For unit above 250 Rs. 1.50/unit<br>
        An additional surcharge of 20% is added to the bill

```python
units = float(input("Enter the total units: "))
total = 0

if units >= 50:
    total += 50 * 0.5
elif units > 0:
    total += units * 0.5

units -= 50

if units >= 100:
    total += 100 * 0.75
elif units > 0:
    total += units * 0.75

units -= 100

if units >= 100:
    total += 100 * 1.2
elif units > 0:
    total += units * 1.2

units -= 100

if units > 0:
    total += units * 1.5

total += total * 0.2
print("Total electricity bill:", total)
```

4.  A company decided to give bonus to employee according to following criteria:<br>
        Period of Service                Bonus<br>
        More than 10 years           	  10%<br>
        >=6 and <=10                 	  8%<br>
        Less than 6 years           	  5%<br>
        Ask user for their salary and years of service and print the net bonus amount.

```python
salary = float(input("Enter the salary: "))
years = int(input("Enter the no. of years: "))

if years > 10:
    bonus = salary * 0.1
elif years > 6:
    bonus = salary * 0.08
elif years >= 0:
    bonus = salary * 0.05
else:
    bonus = 0

print("Net bonus:", bonus)
```

6. Write a program for checking the speed of drivers.<br>
        1.	If speed is less than 70, it should print “Ok”.<br>
        2.	Otherwise, for every 5km above the speed limit (70), it should give the driver one demerit point and print the total number of demerit<br>
            points. For example, if the speed is 80, it should print: “Points: 2”.<br>
        3.	If the driver gets more than 12 points, it should print: “License suspended”

```python
speed = float(input("Enter the speed of the driver: "))

if speed < 70:
    print("OK")
elif speed >= 0:
    points = int(speed - 70) // 5
    print("Points:", points)

    if points > 12:
        print("License Suspended")
else:
    print("Speed cannot be negative")
```

7.  A year is a leap year if it is divisible by 4, except that years divisible by 100 are not leap years unless they are also divisible by 400. Write<br>
    a program that asks the user for a year and prints out whether it is a leap year or not.

```python
year = int(input("Enter the year: "))

if year % 4 == 0 and year % 100 != 0 or year % 400 == 0:
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

9.  Write a program to check whether the triangle is equilateral, isosceles or scalene triangle.

```python
side1 = float(input("Enter the first side of the triangle: "))
side2 = float(input("Enter the second side of the triangle: "))
side3 = float(input("Enter the third side of the triangle: "))

if side1 + side2 > side3 and side2 + side3 > side1 and side3 + side1 > side2:
    if side1 == side2 == side3:
        print("Triangle is Equilateral")
    elif side1 == side2 or side2 == side3 or side3 == side1:
        print("Triangle is Isosceles")
    else:
        print("Triangle is Scalene")
else:
    print("Triangle is not valid")
```

11.  An automobile company manufactures both a two wheeler (TW) and a four wheeler (FW). A company manager wants to make the production of both types<br>
     of vehicle according to the given data below:<br>
        1st data, Total number of vehicle (two-wheeler + four-wheeler)=v<br>
        2nd data, Total number of wheels = W<br>
        The task is to find how many two-wheelers as well as four-wheelers need to manufacture.<br>
        Example:<br>
        Input:<br>
        200 -> Value of V<br>
        540 -> Value of W<br>
        Output:<br>
        TW =130 FW=70

```python
"""
Let, x be the no. of Two-Wheelers that needs to be manufactured
and, y be the no. of Four-Wheelers that needs to be manufactured

According to the question,
=> x + y = v            (I)
=> 2x + 4y = w          (II)

From equation I, we get
=> x = v - y            (III)

Substituting the value of x from equation III into equation II, we get
=> 2(v - y) + 4y = w
=> 2v - 2y + 4y = w
=> 2y = -2v + w
=> y = (w - 2v) / 2
=> y = w / 2 - v        (IV)

Substituting the value of y from equation IV into equation I, we get
=> x = v - (w / 2 - v)
=> x = v - w / 2 + v
=> x = 2v - w / 2       (V)

Therefore, from equation IV and V, we get
x = 2v - w / 2          and,
y = w / 2 - v
"""

v = int(input("Enter total no. vehicles: "))
w = int(input("Enter the total no. of wheels: "))
print("Total Two-Wheeler needs to be manufactured:", 2 * v - w // 2)
print("Total Four-Wheeler needs to be manufactured:", w // 2 - v)
```
