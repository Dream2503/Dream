Q9. Create a 2d array with 1 on the border and 0 inside  
`hint: array[1:-1, 1:-1]`

Q10. Normalize a 5x5 random matrix   
`hint: (x -mean)/std` 

```python
array = np.random.random(5 * 5).reshape(5, 5)
print("Normalized matrix: ", (array - np.mean(array)) / np.std(array), sep='\n')
```

Q11. Multiply a 5x3 matrix by a 3x2 matrix (real matrix product)

```python
array1 = np.random.random(5 * 3).reshape(5, 3)
array2 = np.random.random(3 * 2).reshape(3, 2)
print("Matrix Multiplication of", array1, "and", array2, "is:", array1 @ array2, sep='\n')
```

Q12. Given a 1D array, negate all elements which are between 3 and 8, in place.

```python
array = np.array(eval(input("Enter a array of size 10:")))
array[(array >= 3) & (array <= 8)] *= -1
print("Array after negating elemnts between 3 and 8 are:", array)
```

Q13.  Find the eigenvalues and eigenvectors of a square matrix.  
`hint: np.linalg.eig`

```python
array = np.array(eval(input("Enter a square matrix:")))
eigen_values, eigen_vectors = np.linalg.eig(array)
print("The square matrix is:", array, sep='\n')
print("The eigen values are:", eigen_values)
print("The eigen vectors are:", eigen_vectors, sep='\n')
```

Q14. Find the inverse of a square matrix.

Q25. Delete and Insert a Column in a NumPy Array

```python
array = np.arange(12).reshape(3, 4)
print("The array is:", array, sep='\n')
print("After deleting second column:", np.delete(array, 1, axis=1), sep='\n')
new_col = np.array([[100], [200], [300]])
print(
    "After inserting a new column", new_col, "The resultant array is:", np.insert(array, 1, new_col, axis=1), sep='\n'
    )
```

Q26. Swap column 1 and 2 in a 2D array

```python
array = np.arange(12).reshape(3, 4)
print("After swapping column 1 with column 2 of:", array, sep='\n')
array[:, [0, 1]] = array[:, [1, 0]]
print("The resultant array is:", array, sep='\n')
```

Q29. Sort the rows of a 2D array based on the values of the second column

```python
array = np.array([[1,3],[2,1],[4,2]])
print("The 2D array is:", array, sep='\n')
print("Sorted baseed on second column values:", array[array[:,1].argsort()], sep='\n')
```

Q30. Count the number of occurrences of each unique element

```python
array = np.array([1,2,2,3,3,3])
print("The duplicate elements array is:", array)
unique, counts = np.unique(array, return_counts=True)
print("Number and its frequency is:", dict(zip(unique, counts)))
```

Q31. Transposing a matrix means flipping it over its diagonal.

```python
array = np.array([[1,2],[3,4]])
print("The array is:", array, sep='\n')
print("Its transpose is:", array.T, sep='\n')
```

Q32. Compute the eigenvalues and eigenvectors of a matrix

```python
array = np.array([[1,2],[3,4]])
print("The array is:", array, sep='\n')
values, vectors = np.linalg.eig(array)
print("Eigenvalues:", values)
print("Eigenvectors:", vectors, sep='\n')
```

Q33. Solve a linear equation: x + 2y = 8 and 3x + 4y = 18.

```python
A = np.array([[1,2],[3,4]])
B = np.array([8,18])

solution = np.linalg.solve(A, B).round()
print("Solution of x + 2y = 8 and 3x + 4y = 18 is:")
print("x =", solution[0], "y =", solution[1])
```

Q34. Compute the mean, median, and standard deviation of a NumPy array

```python
array = np.array([1,2,3,4,5])
print("The array is:", array)
print("Mean:", np.mean(array))
print("Median:", np.median(array))
print("Standard Deviation:", np.std(array))
```
