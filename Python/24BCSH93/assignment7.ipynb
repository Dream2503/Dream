Q1. Write a function that returns the sum of digits of a number using recursion.

```python
def sum_of_digits(n):
    if not n:
        return n

    return n % 10 + sum_of_digits(n // 10)


n = int(input("Enter a number: "))
print("Sum of the digits of", n, "are:", sum_of_digits(n))
```

Q2. Write a function that checks whether a number is an Armstrong number (without using built in power function).

```python
from math import log10


def is_armstrong(n):
    count, res, temp = int(log10(n) + 1), 0, n

    while temp:
        digit, power = temp % 10, 1

        for i in range(count):
            power *= digit

        res += power
        temp //= 10

    return n == res


n = int(input("Enter a number: "))
print("Is", n, "an armstrong number?", is_armstrong(n))
```

Q3. Write a function that returns the largest digit in a number (without converting to string).

```python
def largest_digit(n):
    res = 0

    while n:
        res = max(res, n % 10)
        n //= 10

    return res


n = int(input("Enter a number: "))
print("Largest digit in", n, "is:", largest_digit(n))
```

Q4. Write a function that takes a number and returns True if the number is palindrome and False otherwise (without using string conversion).

```python
def is_palindrome(n):
    rev, temp = 0, n

    while temp:
        rev = rev * 10 + temp % 10
        temp //= 10

    return rev == n


n = int(input("Enter a number: "))
print("Is", n, "a palindrome number?", is_palindrome(n))
```

Q5. Write a function that returns the count of prime digits in a number.

```python
def prime_count(n):
    res = 0

    while n:
        res += n % 10 in (2, 3, 5, 7)
        n //= 10

    return res


n = int(input("Enter a number: "))
print("No. of prime digits in", n, "are:", prime_count(n))
```

Q6. Write a function that rotates digits of a number k times to the right.  
    Example: rotate(12345,2) → 45123

```python
def right_rotate(n, k):
    n = str(n)
    k %= len(n)
    return int(n[-k:] + n[:-k])


n = int(input("Enter a number: "))
k = int(input("Enter no. of places to rotate: "))
print("Right rotation of", n, "by", k, "places is:", right_rotate(n, k))
```

Q7. Write a function that returns the factorial of a number using recursion without using loops.

```python
def factorial(n):
    if n <= 1:
        return 1

    return n * factorial(n - 1)


n = int(input("Enter a number: "))
print("Factorial of", n, "is:", factorial(n))
```

Q8. Write a function that checks whether a number is a Perfect Number.

```python
def is_perfect(n):
    res = 1

    for i in range(2, n // 2 + 1):
        if n % i == 0:
            res += i

    return res == n


n = int(input("Enter a number: "))
print("Is", n, "an prefect number?", is_perfect(n))
```

Q9. Write a function that returns the GCD of two numbers using recursion.

```python
def gcd(a, b):
    if b == 0:
        return a

    return gcd(b, a % b)


a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
print("GCD of", a, "and", b, "is:", gcd(a, b))
```

Q10. Write a function that returns the LCM of two numbers using the GCD function.

```python
def gcd(a, b):
    if not b:
        return a

    return gcd(b, a % b)


lcm = lambda a, b: a * b // gcd(a, b)
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
print("LCM of ", a, "and", b, "is:", lcm(a, b))
```

Q11. Write a function that returns the second largest element in a list without using `sort()` or `max()`.

```python
def second_largest(lst):
    res = largest = lst[0]

    for num in lst[1:]:
        if num > largest:
            res = largest
            largest = num

        elif num > res and num != largest:
            res = num

    return res


lst = list(eval(input("Enter a list of numbers: ")))
print("Second largest number in", lst, "is:", second_largest(lst))
```

Q12. Write a function that counts the number of vowels in a string without using built-in count().

```python
def vowel_count(string):
    res = 0

    for ch in string.lower():
        res += ch in ('a', 'e', 'i', 'o', 'u')

    return res


string = input("Enter a string: ")
print("No. of vowels in", string, "are:", vowel_count(string))
```

Q13. Write a function that removes duplicate elements from a list without using `set()`. 

```python
def remove_duplicate(lst):
    res = []

    for num in lst:
        if num not in res:
            res.append(num)

    return res


lst = list(eval(input("Enter a list of numbers: ")))
print("After removing duplicate elements from", lst, "is:", remove_duplicate(lst))
```

Q15. Write a function that checks whether two strings are anagrams (without using sorted()).

```python
def is_anagram(string1, string2):
    lst1, lst2 = list(string1), list(string2)
    lst1.sort()
    lst2.sort()
    return lst1 == lst2


string1 = input("Enter the first string: ")
string2 = input("Enter the second string: ")
print("Is", string1, "and", string2, "are anagrams?", is_anagram(string1, string2))
```

Q16. Write a function that returns the sum of factorial of each digit of a number.

```python
from math import factorial


def factorial_sum(n):
    res = 0

    while n:
        res += factorial(n % 10)
        n //= 10

    return res


n = int(input("Enter a number: "))
print("Factorial of every digit of", n, "is:", factorial_sum(n))
```

Q17. Write a function that returns the nth Fibonacci number using recursion.

```python
def nth_fibonacci(n):
    if n <= 1:
        return n

    return nth_fibonacci(n - 1) + nth_fibonacci(n - 2)


n = int(input("Enter a number: "))
print(n, "th fibonacci number is: ", nth_fibonacci(n), sep="")
```

Q18. Write a function that returns the next greater number formed using the same digits  
     Input: 5342  
     Output: 5423. 

```python
def next_greatest(n):
    res = n + 1
    sort = str(sorted(str(n)))

    if str(n) == sort:
        return n

    while str(sorted(str(res))) != sort:
        res += 1

    return res


n = int(input("Enter a number: "))
print("Next greatest number after", n, "with same digits is:", next_greatest(n))
```
