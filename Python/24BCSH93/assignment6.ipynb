Q1. Write a program to count the number of times a character appears in a given string. 

```python
string = input("Enter a string: ")
freq = {}

for ch in string:
    freq[ch] = freq.get(ch, 0) + 1

print("Frequency of every character is:", freq)
```

Q3. Create a dictionary whose keys are month names and whose values are the number of days in the corresponding months.  
    (a) Ask the user to enter a month name and use the dictionary to tell how many days are in the month.  
    (b) Print out all of the keys in alphabetical order.  
    (c) Print out all of the months with 31 days.  
    (d) Print out the (key-value) pairs sorted by the number of days in each month.  

```python
MAPPING = {
    "January": 31,
    "February": 28,
    "March": 31,
    "April": 30,
    "May": 31,
    "June": 30,
    "July": 31,
    "August": 31,
    "September": 30,
    "October": 31,
    "November": 30,
    "December": 31,
}

name = input("Enter a month name: ").title()
print("(a) Number of days in", name, "are:", MAPPING[name])
print("(b) Keys in alphabetical order:", sorted(MAPPING.keys()))
print("(c) Months with 31 days are:", [name for name, days in MAPPING.items() if days == 31])
print("(d) key-value pairs stored by number of days: ", sorted(MAPPING.items(), key=lambda x: x[1]))
```

Q6. Write a program that checks if two same values in a dictionary have different keys. That is, for dictionary D1 = { 'a' : 10, 'b': 20, 'c' : 10}, the program should print 2 keys have same values and for dictionary D2 = {'a' : 10, 'b' : 20, 'c' : 30}, the program should print No keys have same values. 

```python
d1 = eval(input("Enter a dictionary: "))
d2 = eval(input("Enter another dictionary: "))
found = False
seen = {}

for value in d1.values():
    seen[value] = seen.get(value, 0) + 1

for key, value in seen.items():
    if value > 1:
        if not found:
            print("No. of keys with same value are:")

        found = True
        print("Value", value, "repeated", key, "times")

if not found:
    print("No key have repeted values")
```

Q7. Write a program to check if a dictionary is contained in another dictionary e.g.,  
if  
d1 = {1:11, 2:12}  
d2 = {1:11, 2:12, 3:13, 4:15}  
then d1 is contained in d2.

```python
d1 = eval(input("Enter a dictionary: "))
d2 = eval(input("Enter another dictionary: "))
flag = True

for key, value in d1.items():
    if key not in d2 or d2[key] != value:
        flag = False

if flag:
    print("d2 contains d1")

flag = True

for key, value in d2.items():
    if key not in d1 or d1[key] != value:
        flag = False

if flag:
    print("d1 contains d2")
```

Q8. A dictionary D1 has values in the form of lists of numbers. Write a program to create a new dictionary D2 having same keys as D1 but values as the sum of the list elements e.g.,  
D1 = {'A' : [1, 2, 3] , 'B' : [4, 5, 6]}  
then  
D2 is {'A' :6, 'B' : 15} 

```python
d1 = eval(input("Enter a dictionary: "))
d2 = {key: sum(value) for key, value in d1.items()}
print("Resultant dict:", d2)
```

Q13. Invert Dictionary with Duplicate Value  
    Input: d = {'a':1, 'b':2, 'c':1, 'd':3, 'e':2}  
    Output: {1: ['a', 'c'], 2: ['b', 'e'], 3: ['d']}  

```python
d = eval(input("Enter a dictionary: "))
res = {}

for key, value in d.items():
    res[value] = res.get(value, [])
    res[value].append(key)

print("resultant dict:", res)
```

Q14. Using dictionary, print the first character that does not repeat.  
    Input: "aabbccddeefg"

```python
string = input("Enter a string: ")
freq = {}

for ch in string:
    freq[ch] = freq.get(ch, 0) + 1

for ch in string:
    if freq[ch] == 1:
        print("The first character that does not repeat:", ch)
        break
```

Q16. Create a dictionary where keys are word lengths and values are lists of words.  
        Input: ["cat","dog","elephant","rat","tiger"]  
        Output: {3:['cat','dog','rat'], 5:['tiger'],8:['elephant']}  

```python
words = list(eval(input("Enter the list of strings: ")))
dict1 = {}

for word in words:
    dict1[len(word)] = dict1.get(len(word), [])
    dict1[len(word)].append(word)

print("Resultant dict is:", dict1)
```
