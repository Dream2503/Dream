Q17. Create a DataFrame with columns: Roll No, Name, Marks1, Marks2, and Marks3. Perform the following operations:
     1.	Calculate Total Marks and Percentage
     2.	Add a new column Grade based on percentage
     3.	Display students scoring more than 75%
     4.	Find the student with second highest percentage
     5.	Sort the DataFrame based on percentage

```python
df = pd.DataFrame(
        {
            "Roll No": [1, 2, 3, 4, 5],
            "Name": ["A", "B", "C", "D", "E"],
            "Marks1": [80, 70, 90, 60, 85],
            "Marks2": [75, 65, 95, 55, 80],
            "Marks3": [85, 60, 92, 58, 88]
        }
)

df["Total"] = df[["Marks1", "Marks2", "Marks3"]].sum(axis=1)
df["Percentage"] = df["Total"] / 3


def grade(p):
    if p >= 75:
        return "A"
    elif p >= 60:
        return "B"
    else:
        return "C"


df["Grade"] = df["Percentage"].apply(grade)
print("Student details are:", df, sep='\n')
print("Students with more than 75% marks are:", df[df["Percentage"] > 75], sep='\n')
print("Second highest marks:", df.sort_values("Percentage", ascending=False).iloc[1], sep='\n')
print("Sorted DataFrame:", df.sort_values("Percentage", ascending=False), sep='\n')
```

Q18. Create a DataFrame with fields: EmpID, Name, Department, and Salary. Write a program to:
     1.	Display employees of a specific department (user input)
     2.	Increase salary by 10% for all employees
     3.	Find the highest and lowest salary
     4.	Group employees by department and compute average salary
     5.	Save the DataFrame to a CSV file

```python
employees = pd.DataFrame(
        {
            "EmpID": [1, 2, 3, 4],
            "Name": ["A", "B", "C", "D"],
            "Department": ["IT", "HR", "IT", "Finance"],
            "Salary": [50000, 40000, 60000, 45000]
        }
)
print("Employees details are:", employees, sep='\n')
print("The employees of IT department are:", employees[employees["Department"] == "IT"], sep='\n')
employees["Salary"] *= 1.1
print("Employees after increase on salary:", employees, sep='\n')
print("Maximum salary of employees:", employees["Salary"].max())
print("Minimum salary of employees:", employees["Salary"].min())
print("Average salary department wise:", employees.groupby("Department")["Salary"].mean(), sep='\n')
employees.to_csv("employees.csv", index=False)
```

Q19. Create a DataFrame student with columns: roll No, name, marks, and city, perform:
     1.	Select rows where marks > 80
     2.	Filter students from a specific city
     3.	Select specific columns using loc[] and iloc[]
     4.	Update marks of a particular student
     5.	Delete a column

```python
student = pd.DataFrame(
        {
            "roll no": [1, 2, 3],
            "name": ["A", "B", "C"],
            "marks": [85, 70, 90],
            "city": ["Delhi", "Mumbai", "Delhi"]
        }
)
print("Student details are:", student, sep='\n')
print("Students with marks >80:", student[student["marks"] > 80], sep='\n')
print("Students from delhi:", student[student["city"] == "Delhi"], sep='\n')
print("Use of loc:", student.loc[:, ["name", "marks"]], sep='\n')
print("Use of iloc:", student.iloc[:, 1:3], sep='\n')
student.loc[student["roll no"] == 2, "marks"] = 75
print("After updating the marks of roll no. 2:", student, sep='\n')
student = student.drop("city", axis=1)
print("After deleting city column:", student, sep='\n')
```

Q20. A company stores product sales data with columns: ProductID, ProductName, Category, Price, and Quantity. Write a program to:
     1.	Calculate total revenue
     2.	Find top-selling products
     3.	Filter products with low sales
     4.	Group by category and compute total revenue
     5.	Export results to Excel

```python
sales = pd.DataFrame(
        {
            "ProductID": [1, 2, 3],
            "ProductName": ["A", "B", "C"],
            "Category": ["X", "Y", "X"],
            "Price": [100, 200, 150],
            "Quantity": [5, 2, 8]
        }
)
sales["Revenue"] = sales["Price"] * sales["Quantity"]
print("Total revenue is:", sales["Revenue"].sum())
print("Top selling products:", sales.sort_values("Quantity", ascending=False), sep='\n')
print("Low sale products are:", sales[sales["Quantity"] < 3], sep='\n')
print("Total revenue category-wise:", sales.groupby("Category")["Revenue"].sum())
sales.to_excel("sales.xlsx", index=False)
```

Q21. A CSV file students.csv contains: RollNo, Name, Marks1, Marks2, and Marks3. Write a Python program to:
     1.	Read the CSV file into a DataFrame
     2.	Calculate Total and Percentage
     3.	Assign grades based on percentage
     4.	Display students scoring above 75%
     5.	Find the student with second highest percentage
     6.	Save the result into result.csv

```python
df = pd.read_csv("students.csv")
print("Student details are:", df, sep='\n')
df["Total"] = df[["Marks1","Marks2","Marks3"]].sum(axis=1)
df["Percentage"] = df["Total"] / 3
print("After calculating total and percentage:", df, sep='\n')

def grade(p):
    if p >= 75:
        return "A"
    elif p >= 60:
        return "B"
    else:
        return "C"

df["Grade"] = df["Percentage"].apply(grade)
print("After assigning the grades:", df, sep='\n')
print("Students scoring more than 75%:", df[df["Percentage"] > 75], sep='\n')
print("Second highest mark:", df.sort_values("Percentage", ascending=False).iloc[1])
df.to_csv("result.csv", index=False)
```
