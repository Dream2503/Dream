Q1. Create a null vector of size 10 but the fifth value is 1.

```python
array = np.zeros(10)
array[4] = 1
print("Resultant vector is:", array)
```

Q2. Create a vector with values ranging from 10 to 49

Q3. Reverse a vector (first element becomes last) 

```python
array = np.array(eval(input("Enter a vector: ")))
print("The reverse of", array, "is:", array[::-1])
```

Q6. Create a 3x3x3 array with random values  
`hint: np.random.random`

```python
print("3x3x3 random array is:", np.random.random(3 * 3 * 3).reshape(3, 3, 3), sep='\n')
```

Q7. Create a 10x10 array with random values and find the minimum and maximum values   
`hint: min, max`

```python
array = np.random.random(10 * 10).reshape(10, 10)
print("The randon 10x10 array is:", array, sep='\n')
print("Min element in array is:", np.min(array))
print("Max element in array is:", np.max(array))
```

Q8. Create a random vector of size 30 and find the mean value  
`hint: mean` 

```python
array = np.random.random(30)
print("The random vector of size 30 is:", array)
print("The mean of the vector is:", np.mean(array))
```

Q9. Create a 2d array with 1 on the border and 0 inside  
`hint: array[1:-1, 1:-1]`

```python
row, column = eval(input("Enter the dimension of the matrix"))
array = np.ones(row * column).reshape(row, column)
array[1:-1, 1:-1] = 0
print(array)
```

Q10. Normalize a 5x5 random matrix   
`hint: (x -mean)/std` 

```python
array = np.random.random(5 * 5).reshape(5, 5)
print("Normalized matrix: ", (array - np.mean(array)) / np.std(array), sep='\n')
```

Q11. Multiply a 5x3 matrix by a 3x2 matrix (real matrix product)

```python
array1 = np.random.random(5 * 3).reshape(5, 3)
array2 = np.random.random(3 * 2).reshape(3, 2)
print("Matrix Multiplication of", array1, "and", array2, "is:", array1 @ array2, sep='\n')
```

Q12. Given a 1D array, negate all elements which are between 3 and 8, in place.

```python
array = np.array(eval(input("Enter a array of size 10:")))
array[(array >= 3) & (array <= 8)] *= -1
print("Array after negating elemnts between 3 and 8 are:", array)
```

Q13.  Find the eigenvalues and eigenvectors of a square matrix.  
`hint: np.linalg.eig`

```python
array = np.array(eval(input("Enter a square matrix:")))
eigen_values, eigen_vectors = np.linalg.eig(array)
print("The square matrix is:", array, sep='\n')
print("The eigen values are:", eigen_values)
print("The eigen vectors are:", eigen_vectors, sep='\n')
```

Q14. Find the inverse of a square matrix.  
`hint: np.linalg.inv`

```python
array = np.array(eval(input("Enter a square matrix:")))
print("Inverse of", array, "is:", np.linalg.inv(array), sep='\n')
```
