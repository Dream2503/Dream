Q9. Split a file into multiple files of equal number of lines.

```python
path = input("Enter the file name: ")
lines_count = int(input("Enter the no. of lines in each file: "))

try:
    with open(path, 'r') as f1:
        lines = f1.readlines()

        if len(lines) % lines_count != 0:
            raise Exception("Cannot be split into equal no. of lines")

        for i in range(len(lines) // lines_count):
            with open(f"split{i}.txt", 'w') as f2:
                f2.writelines(lines[i * lines_count: (i + 1) * lines_count])

        print("Successfully split across", len(lines) // lines_count, "files")

except Exception as e:
    print(e)
```

Q10. Print duplicate lines from a file.

```python
path = input("Enter the file name: ")
seen = set()
duplicate = set()

try:
    with open(path, 'r') as f:
        for line in f.readlines():
            if line in seen:
                duplicate.add(line)

            else:
                seen.add(line)

    print("Duplicate lines are:")
    print(*duplicate, sep='\n')

except Exception as e:
    print(e)
```

Q28. Bank Minimum Balance
Create a Python program that implements a Bank Account class with methods for deposit and withdraw operations. The account must maintain a minimum balance of Rs.1000. If the withdrawal amount violates the minimum balance rule, the program should raise and handle a user-defined exception.

```python
class MinimumBalanceError(Exception):
    pass


class BankAccount:
    MIN_BALANCE = 1000

    def __init__(self, account_holder):
        self.account_holder = account_holder
        self.balance = 0

    def deposit(self, amount):
        if amount < 0:
            print("Amount cannot be negative")
            return

        self.balance += amount

    def withdraw(self, amount):
        if amount < 0:
            print("Amount cannot be negative")
            return

        if self.balance - amount < self.MIN_BALANCE:
            raise MinimumBalanceError()

        self.balance -= amount

    def display_balance(self):
        print("Balance:", self.balance)


try:
    account_holder = input("Enter the name of the account holder: ")
    bank_account = BankAccount(account_holder)
    amount = float(input("Enter the amount to deposit: "))
    bank_account.deposit(amount)
    amount = float(input("Enter the amount to withdraw: "))
    bank_account.withdraw(amount)
    bank_account.display_balance()

except MinimumBalanceError as e:
    print("Exception:", e)
```

Q29. Product Stock Management
Write a Python program for a product inventory system If the user tries to purchase more items than available in stock, raise a user-defined exception called OutOfStockError.

```python
class OutOfStockError(Exception):
    pass


class Product:
    def __init__(self, product_name):
        self.product_name = product_name
        self.stock = 0

    def add_stock(self, quantity):
        if quantity < 0:
            print("Quantity cannot be negative")
            return

        self.stock += quantity

    def purchase(self, quantity):
        if quantity < 0:
            print("Quantity cannot be negative")
            return

        if quantity > self.stock:
            raise OutOfStockError()

        self.stock -= quantity

    def display_stock(self):
        print("Available stock:", self.stock)


try:
    product_name = input("Enter product name: ")
    product = Product(product_name)
    quantity = int(input("Enter quantity to add to stock: "))
    product.add_stock(quantity)
    quantity = int(input("Enter quantity to purchase: "))
    product.purchase(quantity)
    product.display_stock()

except OutOfStockError as e:
    print("Exception:", e)
```

```python
from sqlite3 import connect

with connect("college.db") as connection:
    cursor = connection.cursor()
    cursor.execute(
            """
            CREATE TABLE IF NOT EXISTS student (
                roll_no INT PRIMARY KEY,
                name    VARCHAR(30),
                marks   NUMERIC(5, 2)
            );
            """,
    )
    print("Enter 3 student records (Roll No., Name, Marks)")

    for i in range(1, 4):
        data = eval(input(f"Enter the details of Student{i}:"))
        cursor.execute(
                """
                INSERT INTO student
                VALUES (?, ?, ?);
                """, data,
        )
    print("The content of the table is:")

    for row in cursor.execute(
            """
            SELECT *
            FROM student;
            """,
    ):
        print(row)

    mark = float(input("Enter the updated marks of Roll No. 1: "))
    cursor.execute(
            """
            UPDATE student
            SET marks = ?
            WHERE roll_no = 1;
            """, (mark,),
    )
    roll = int(input("Enter the Roll No. to delete the record: "))
    cursor.execute(
            """
            DELETE
            FROM student
            WHERE roll_no = ?;
            """, (roll,),
    )
    print("A record from the table:")

    print(
            cursor.execute(
                    """
                    SELECT *
                    FROM student
                    LIMIT 1;
                    """, ).fetchone()
    )
    print(
            "Total no. of records in the table is:", *cursor.execute(
                    """
                    SELECT COUNT(*)
                    FROM student;
                    """,
            ).fetchone()
    )
    cursor.close()
```

Q37. Student Database System
Develop a Python program to:
Create a database school.db

RollNo, Name, Marks
Insert at least 5 records

All students
Students scoring above average marks

```python
from sqlite3 import connect

with connect("school.db") as connection:
    cursor = connection.cursor()

    cursor.execute(
            """
            CREATE TABLE IF NOT EXISTS student (
                roll_no INT PRIMARY KEY,
                name    VARCHAR(30),
                marks   NUMERIC(5, 2)
            );
            """
    ).executemany(
            """
            INSERT INTO student
            VALUES (?, ?, ?);
            """, [
                (1, "Alice", 82),
                (2, "Bob", 67),
                (3, "Charlie", 91),
                (4, "David", 74),
                (5, "Eve", 88)
            ]
    )
    print("All Students:")

    for row in cursor.execute(
            """
            SELECT *
            FROM student;
            """
    ):
        print(row)

    print("\nStudents scoring above average marks:")
    for row in cursor.execute(
            """
            SELECT *
            FROM student
            WHERE marks > (SELECT AVG(marks)
                           FROM student);
            """
    ):
        print(row)

    cursor.close()
```

Q38. Employee Management System
Create a class-based Python program to:
Store employee details in SQLite database:
empID, empName, Salary
Perform operations:
Insert
Update
Delete
Display employees with salary > 50,000

```python
from sqlite3 import connect


class EmployeeDB:
    def __init__(self):
        self.connection = connect("company.db")
        self.cursor = self.connection.cursor()
        self.cursor.execute(
                """
                CREATE TABLE IF NOT EXISTS employee (
                    empID   INTEGER PRIMARY KEY,
                    empName TEXT,
                    Salary  REAL
                );
                """
        )
        self.connection.commit()

    def insert_employee(self, empID, empName, salary):
        self.cursor.execute(
                """
                INSERT INTO employee
                VALUES (?, ?, ?)
                """, (empID, empName, salary)
        )
        self.connection.commit()

    def update_salary(self, empID, new_salary):
        self.cursor.execute(
                """
                UPDATE employee
                SET Salary = ?
                WHERE empID = ?
                """, (new_salary, empID)
        )
        self.connection.commit()

    def delete_employee(self, empID):
        self.cursor.execute(
                """
                DELETE
                FROM employee
                WHERE empID = ?
                """, (empID,)
        )
        self.connection.commit()

    def display_all(self):
        print("\nAll Employees:")

        for row in self.cursor.execute(
                """
                SELECT *
                FROM employee
                """
        ):
            print(row)

    def display_high_salary(self):
        print("\nEmployees with salary > 50000:")

        for row in self.cursor.execute(
                """
                SELECT *
                FROM employee
                WHERE Salary > 50000
                """
        ):
            print(row)

    def close(self):
        self.connection.close()


db = EmployeeDB()
db.insert_employee(1, "Alice", 45000)
db.insert_employee(2, "Bob", 60000)
db.insert_employee(3, "Charlie", 75000)
db.update_salary(1, 52000)
db.delete_employee(3)
db.display_all()
db.display_high_salary()
db.close()
```
