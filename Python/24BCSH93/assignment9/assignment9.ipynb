Q1. Write a Python program to count the number of lines, words, and characters in a file.

```python
path = input("Enter the file name: ")

try:
    with open(path, 'r') as f:
        content = f.read()
    
    print("No. of characters:", len(content))
    print("No. of words:", len(content.split()))
    print("No. of lines:", content.count('\n') + 1)
    
except Exception as e:
    print(e)
```

Q2. Copy contents from one file to another file.

```python
from_path= input("Enter the file name to copy from: ")
to_path = input("Enter the file name to copy to: ")

try:
    with open(from_path, 'r') as f1, open(to_path, 'w') as f2:
        f2.write(f1.read())

    print("Content of", from_path, "copied to", to_path)

except Exception as e:
    print(e)
```

Q3. Display only those lines from a file that start with a vowel.

```python
path = input("Enter the file name: ")

try:
    with open(path, 'r') as f:
        print("Lines which starts with vowel are:")
        print(*(line for line in f.readlines() if line[0].lower() in "aeiou"))
    
except Exception as e:
    print(e)
```

Q4. Count how many times each word appears in a file.

```python
path = input("Enter the file name: ")

try:
    with open(path, 'r') as f:
        words = f.read().split()
        print("Frequency of words are:")
        print(*({word : words.count(word) for word in words}.items()), sep='\n')
    
except Exception as e:
    print(e)
```

Q5. Search for a word and display the line numbers where it appears.

```python
path = input("Enter the file name: ")
key = input("Enter the word to search: ")

try:
    with open(path, 'r') as f:
        lines = f.readlines()
        
    for i in range(len(lines)):
        if key in lines[i]:
            print(key, "found at line no.", i + 1)
            break
    
except Exception as e:
    print(e)
```

Q6. Store student details in a file and display students scoring more than 75 marks.

```python
students = eval(input("Enter the name and marks of the students: "))
file_name = "students.txt"
try:
    with open(file_name, 'w') as f:
        for student in students:
            f.write(f"{student[0]},{student[1]}\n")
            
except Exception as e:
    print(e)
    
print("Students scoring more than 75:")

try:
    with open(file_name, 'r') as f:
        for line in f.readlines():
            name, marks = line[:-1].split(',')
            
            if int(marks) > 75:
                print(f"{name}: {marks}")
                
except Exception as e:
    print(e)
```

Q7. Count uppercase letters, lowercase letters, and digits in a file.

```python
path = input("Enter the file name: ")
count = [0] * 3

try:
    with open(path, 'r') as f:
        for ch in f.read():
            count[0 if ch.isupper() else 1 if ch.islower() else 2] += 1
    
    print("No. of uppercase letters:", count[0])
    print("No. of lowercase letters:", count[1])
    print("No. of digits:", count[2])
    
except Exception as e:
    print(e)
```

Q8. Find the longest word in a file

```python
path = input("Enter the file name: ")

try:
    with open(path, 'r') as f:
        print("The longest word:", max(*f.read().split(), key=lambda x: len(x)))
    
except Exception as e:
    print(e)
```

Q9. Split a file into multiple files of equal number of lines.

```python
path = input("Enter the file name: ")
lines_count = int(input("Enter the no. of lines in each file: "))

try:
    with open(path, 'r') as f1:
        lines = f1.readlines()
        
        if len(lines) % lines_count != 0:
            raise Exception("Cannot be split into equal no. of lines")
        
        for i in range(len(lines) // lines_count):
            with open(f"split{i}.txt", 'w') as f2:
                f2.writelines(lines[i * lines_count: (i + 1) * lines_count])
        
        print("Successfully splited across", len(lines) // lines_count, "files")
    
except Exception as e:
    print(e)
```

Q10. Print duplicate lines from a file.

```python
path = input("Enter the file name: ")
seen = set()
duplicate = set()

try:
    with open(path, 'r') as f:
        for line in f.readlines():
            if line in seen:
                duplicate.add(line)
                
            else:
                seen.add(line)
    
    print("Duplicate lines are:")
    print(*duplicate, sep='\n')
    
except Exception as e:
    print(e)
```

Q11. Create a file containing numbers. Read the file and:  
• Count even and odd numbers  
• Find sum of all numbers 

```python
path = input("Enter the file name: ")
even = odd = sum = 0

try:
    with open(path, 'r') as f:
        numbers = eval(f.read())
    
    for num in numbers:
        sum += num
        
        if num % 2 == 0:
            even += 1
        
        else:
            odd += 1
    
    print("No. of even numbers are:", even)
    print("No. of odd numbers are:",  odd)
    print("Sum of the numbers are:",  sum)
    
except Exception as e:
    print(e)
```

Q12. Write a program to:  
• Read a file  
• Display only those words whose length is greater than 5 

```python
path = input("Enter the file name: ")

try:
    with open(path, 'r') as f:
        print("Words with len greater than 5 are:", *(word for word in f.read().split() if len(word) > 5))
    
except Exception as e:
    print(e)
```
