Q1. Given a range of first 10 numbers, Iterate from start number to the end number and print the sum of the current number and previous number.

```python
for i in range(10):
    if i > 0:
        print("Sum at index", i, ":", 2 * i - 1)
    else:
        print("Sum at index", i, ":", i)
```

Q7. Given a two list of numbers create a new list such that new list should contain only odd numbers from the first list and even numbers from the second list.

Q8. Given a list iterate it and display numbers which are divisible by 5 and if you find number greater than 150 stop the loop iteration.

```python
lst = list(eval(input("Enter a list of numbers(comma-separated): ")))
print("Numbers divisible by 5 are:", end=" ")

for element in lst:
    if element > 150:
        break

    if element % 5 == 0:
        print(element, end=" ")
```

Q9. Use a loop to display elements from a given list which are present at even positions.

Q10.  Find the sum of the series 2 + 22 + 222 + 2222 + ... n terms.

```python
n = int(input("Enter a number: "))
res = current = 0

for i in range(n):
    current = current * 10 + 2
    res += current

print("The sum of the series is:", res)
```

Q12. Given an input list removes the element at index 4 and add it to the 2nd position and also, at the end of the list.

```python
lst = list(eval(input("Enter a list of numbers(comma-separated): ")))

if len(lst) >= 5:
    element = lst.pop(4)
    lst.insert(2, element)
    lst.append(element)
    print("Resultant list:", lst)
else:
    print("Length of the list must be greater than 4")
```

Q13. Given a list slice it into a 3 equal chunks and reverse each list.

```python
lst = list(eval(input("Enter a list of numbers(comma-separated): ")))
size = len(lst) // 3
chunks = [lst[: size], lst[size: size * 2], lst[size * 2:]]

for chunk in chunks:
    chunk.reverse()

print("The 3 equal chunks are:", chunks)
```

Q18. Arrange string characters such that lowercase letters should come first.

```python
string = input("Enter a string: ")
res = ""

for ch in string:
    if ch.islower():
        res = ch + res
    else:
        res += ch

print("Resultant string is:", res)
```

Q19. Count all lower case, upper case, digits, and special symbols from a given string.

```python
string = input("Enter a string: ")
counts = [0] * 4

for ch in string:
    if ch.islower():
        counts[0] += 1
    elif ch.isupper():
        counts[1] += 1
    elif ch.isdigit():
        counts[2] += 1
    else:
        counts[3] += 1

print("Number of lower case letters are:", counts[0])
print("Number of upper case letters are:", counts[1])
print("Number of digits are:", counts[2])
print("Number of special symbols are:", counts[3])
```

Q20. Given a string, return the sum and average of the digits that appear in the string, ignoring all other characters.

```python
string = input("Enter a string: ")
res = cnt = 0

for ch in string:
    if ch.isdigit():
        res += int(ch)
        cnt += 1

print("Sum:", res)
print("Average:", res / cnt)
```
