1.	Write a program to find the smallest and largest digit in a number.

```python
num = int(input("Enter a number: "))
min, max = 9, 0

while num:
    digit = num % 10
    
    if digit > max:
        max = digit
    if digit < min:
        min = digit
    
    num //= 10

print("Smallest digit:", min)
print("Largest Digit:", max)
```

2.	Write a program to find the frequency of each digit in a number.

```python
num = int(input("Enter a number: "))

for i in range(10):
    count, temp = 0, num
    
    while temp:
        if temp % 10 == i:
            count += 1
        
        temp //= 10
    
    if count:
        print("Frequency of", i, "is:", count)
```

3.	Write a program to check whether a number is binary or not.

```python
num = int(input("Enter a number: "))
flag = True

while num:
    if num % 10 not in (0, 1):
        flag = False
    
    num //= 10

if flag:
    print("The number is a bianry number")
else:
    print("The number is not a binary number")
```

4.	Write a program to count how many digits are even and odd in a number.

```python
num = int(input("Enter a number: "))
even = odd = 0

while num:
    if num % 10 % 2 == 0:
        even += 1
    else:
        odd += 1
    
    num //= 10

print("The number of even digits are:", even)
print("The number of odd digits are:", odd)
```

7.	Write a Python program to construct the following pattern, using a nested for loop.
    *
    * * 
    * * * 
    * * * * 
    * * * * * 
    * * * * 
    * * * 
    * * 
    *


```python
num = int(input("Enter a number: "))

for i in range(num):
    for j in range(i + 1):
        print("*", end=" ")
    print()

for i in range(num - 2, -1, -1):
    for j in range(i + 1):
        print("*", end=" ")
    print()
```

9.	Count numbers between 1 and 100 whose reverse is greater than the original number.

```python
count = 0

for i in range(1, 101):
    rev, temp = 0, i
    
    while temp:
        rev = rev * 10 + i % 10
        temp //= 10

    if rev > i:
        count += 1

print("Number of numbers whose reverse is greater than the original number are:", count)
```

10.	Print all numbers between 1 and 100 where product of digits > sum of digits.

```python
count = 0

for i in range(1, 101):
    product, sum, temp = 1, 0, i

    while temp != 0:
        product *= temp % 10
        sum += temp % 10
        temp //= 10
    
    if product > sum:
        count += 1

print("Number of numbers whose product of digits is greater than sum of digits are:", count)
```

13.	 Swap the first and last digit of a number.

```python
num = int(input("Enter a number: "))
res = num

if num > 9:
    last, first, rev, zeroes = num % 10, num // 10, 0, 0

    while first > 9:
        digit = first % 10
        rev = rev * 10 + digit
        first //= 10

        if digit == 0:
            zeroes += 1

    res = last

    while rev:
        digit = rev % 10
        res = res * 10 + digit
        rev //= 10

        if digit == 0:
            zeroes -= 1
    
    while zeroes:
        res *= 10
        zeroes -= 1
        
    res = res * 10 + first

print(res)
```

15.	Check whether a number is a Harshad number (divisible by sum of digits).

```python
num = int(input("Enter a number: "))
sum, temp = 0, num

while temp:
    sum += temp % 10
    temp //= 10

if num % sum == 0:
    print(num, "is a Harshad number")
else:
    print(num, "is not a Harshad number")
```

16.	Find numbers whose square ends with the same digits as the number itself.(Hit: 5->5^2=25)

```python
low = int(input("Enter the low: "))
high = int(input("Enter the high: "))
print("The numbers whose square ends with the same digits as the number are:")

for i in range(low, high + 1):
    squared, temp, flag = i ** 2, i, True

    while temp:
        if squared % 10 != temp % 10:
            flag = False
            break

        temp //= 10
    
    if flag:
        print(i, end=" ")
```

20.	Find the LCM and GCD of two numbers.

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
a, b = num1, num2

while b:
    a, b = b, a % b

print("LCM of", num1, "and", num2, "is:", num1 * num2 // a)
print("GCD of", num1, "and", num2, "is:", a)
```
