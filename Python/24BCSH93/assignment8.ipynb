Q2. Rectangle Area  
    Create a class Rectangle with:  
        • radius  
        • width  
    Add methods:  
        • area() → returns area  
        • perimeter() → returns perimeter

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)


length = float(input("Enter the length: "))
width = float(input("Enter the width: "))
rectangle = Rectangle(length, width)
print("Area:", rectangle.area())
print("Perimeter:", rectangle.perimeter())
```

Q4. Circle Class  
    Create a class Circle with attribute radius.  
    Add methods:  
        • area()  
        • circumference()  
    Use the value π = 3.14.

```python
class Circle:
    PI = 3.14

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return Circle.PI * self.radius ** 2

    def perimeter(self):
        return 2 * Circle.PI * self.radius


radius = float(input("Enter the radius: "))
circle = Circle(radius)
print("Area:", circle.area())
print("Perimeter:", circle.perimeter())
```

        • account_holder  
        • balance  
    Methods:  
        • deposit(amount)

        • display_balance()  
    Add a condition so withdrawal cannot exceed balance. 

```python
class BankAccount:
    def __init__(self, account_holder):
        self.account_holder = account_holder
        self.balance = 0

    def deposit(self, amount):
        if amount < 0:
            print("Amount cannot be negetive")
            exit(1)

        self.balance += amount

    def withdraw(self, amount):
        if amount < 0:
            print("Amount cannot be negetive")
            exit(1)

        if self.balance < amount:
            print("Insufficient Funds")
            exit(1)

        self.balance -= amount

    def display_balance(self):
        print("Balance:", self.balance)


account_holder = input("Enter the name of the account holder: ")
bank_account = BankAccount(account_holder)
amount = float(input("Enter the amount to deposit: "))
bank_account.deposit(amount)
amount = float(input("Enter the amount to wthdraw: "))
bank_account.withdraw(amount)
bank_account.display_balance()
```

Q8. Temperature Converter  
    Create a class Temperature with attribute celsius.  
    Methods:  
        • to_fahrenheit()  
        • to_kelvin()

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    def to_fahrenheit(self):
        return 9 / 5 * self.celsius + 32

    def to_kelvin(self):
        return 273.15 + self.celsius


celsius = float(input("Enter temperature in celsius: "))
temperature = Temperature(celsius)
print("Fahrenheit:", temperature.to_fahrenheit())
print("Kelvin:", temperature.to_kelvin())
```

Q10. Shopping Cart  
     Create a class Cart with:  
        • items (dictionary)  
     Methods:  
        • add_item(item, price)  
        • remove_item(item)  
        • total_price()

```python
class Cart:
    def __init__(self):
        self.items = {}

    def add_item(self, item, price):
        self.items[item] = price

    def remove_item(self, item):
        if item in self.items:
            self.items.pop(item)
        else:
            print(item, "not present in cart")

    def total_price(self):
        return sum(self.items.values())


cart = Cart()
element = eval(input("Enter the item and price to add: "))
cart.add_item(element[0], element[1])
item = input("Enter the item to remove: ")
cart.remove_item(item)
print("Total price:", cart.total_price())
```

Q11. Student Grade System  
     Create a class Student with:  
        • name  
        • marks (list)  
     Methods:  
         • average_marks()  
         • grade()  
         • high_grad()  
     Grades:  
         • A → 90+  
         • B → 75+  
         • C → 50+  
         • F → below 50

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def average_marks(self):
        return sum(self.marks) / len(self.marks)

    def grade(self):
        avg = self.average_marks()

        if avg >= 90:
            return 'A'

        elif avg >= 75:
            return 'B'

        elif avg >= 50:
            return 'C'

        else:
            return 'F'

    def high_grad(self):
        return max(self.marks)

    def display(self):
        print(f"Student(name={self.name}, average={self.average_marks():.2f}, grade={self.grade()}, highest={self.high_grad()})")


name = input("Enter student name: ")
n = int(input("Enter number of subjects: "))
marks = [float(input(f"Enter mark for subject {i + 1}: ")) for i in range(n)]
student = Student(name, marks)
student.display()
```

Q12. ATM Simulation  
     Create a class ATM with:  
     • balance  
     • pin  
     Methods:  
     • check_balance()  
     • deposit()  
     • withdraw()  
     • change_pin()

```python
class ATM:
    def __init__(self, balance, pin):
        self.balance = balance
        self.pin = pin

    def check_balance(self):
        print("Balance:", self.balance)

    def deposit(self, amount):
        self.balance += amount
        print("New balance:", self.balance)

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance.")
            exit(1)

        self.balance -= amount
        print("New balance:", self.balance)

    def change_pin(self, old_pin, new_pin):
        if old_pin != self.pin:
            print("Incorrect PIN.")
            exit(1)

        self.pin = new_pin
        print("PIN changed successfully.")


balance = float(input("Enter initial balance: "))
pin = int(input("Set your PIN: "))
atm = ATM(balance, pin)
attempt = int(input("Enter PIN: "))

if attempt != atm.pin:
    print("Incorrect PIN. Access denied.")
    exit(1)

atm.check_balance()
amount = float(input("Enter amount to deposit: "))
atm.deposit(amount)
amount = float(input("Enter amount to withdraw: "))
atm.withdraw(amount)

old = int(input("Enter old PIN: "))
new = int(input("Enter new PIN: "))
atm.change_pin(old, new)
```

Q13. Stack Data Structure Class with Display Method  
     Write a Python program to create a class representing a stack data structure. Include methods for pushing, popping and displaying elements. 

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if len(self.items) == 0:
            print("Stack is empty.")
            exit(1)

        self.items.pop()

    def display(self):
        print("Stack:", self.items)


stack = Stack()
n = int(input("Enter number of items to push: "))

for i in range(n):
    stack.push(int(input("Enter item: ")))

stack.display()
n = int(input("Enter number of items to pop: "))

for i in range(n):
    stack.pop()

stack.display()
```

Q14.Person Class with Age Calculation  
    Create a Person class with attributes like name and date_of_birth . Implement a method that calculates the person's current age based on the date of birth.

```python
class Person:
    def __init__(self, name, day, month, year):
        self.name = name
        self.day = day
        self.month = month
        self.year = year

    def age(self, current_day, current_month, current_year):
        years = current_year - self.year

        if (current_month, current_day) < (self.month, self.day):
            years -= 1

        return years

    def display(self, current_day, current_month, current_year):
        print(f"Person(name={self.name}, age={self.age(current_day, current_month, current_year)})")


name = input("Enter name: ")
day = int(input("Enter birth day: "))
month = int(input("Enter birth month: "))
year = int(input("Enter birth year: "))

current_day = int(input("Enter current day: "))
current_month = int(input("Enter current month: "))
current_year = int(input("Enter current year: "))

person = Person(name, day, month, year)
person.display(current_day, current_month, current_year)
```

Q16. Define a class Point, which can be used to create two-dimensional points. Create three objects of Point class and check, these points are non collinear.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def non_collinear(self, p2, p3):
        area = self.x * (p2.y - p3.y) + p2.x * (p3.y - self.y) + p3.x * (self.y - p2.y)
        if area == 0:
            print("Points are collinear.")
        else:
            print("Points are non-collinear.")


p1 = Point(int(input("Enter x1: ")), int(input("Enter y1: ")))
p2 = Point(int(input("Enter x2: ")), int(input("Enter y2: ")))
p3 = Point(int(input("Enter x3: ")), int(input("Enter y3: ")))
p1.non_collinear(p2, p3)
```

Q18. Define a class ComplexNumber, placed the necessary data members and member functions to perform basic Complex number operation.

```python
class ComplexNumber:
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary

    def add(self, other):
        return ComplexNumber(self.real + other.real, self.imaginary + other.imaginary)

    def subtract(self, other):
        return ComplexNumber(self.real - other.real, self.imaginary - other.imaginary)

    def multiply(self, other):
        return ComplexNumber(self.real * other.real - self.imaginary * other.imaginary,
                             self.real * other.imaginary + self.imaginary * other.real)

    def display(self):
        print(f"{self.real} + {self.imaginary}i")


c1 = ComplexNumber(float(input("Enter real part of first number: ")), float(input("Enter imaginary part of first number: ")))
c2 = ComplexNumber(float(input("Enter real part of second number: ")), float(input("Enter imaginary part of second number: ")))
print("Addition:", end=" ")
c1.add(c2).display()
print("Subtraction:", end=" ")
c1.subtract(c2).display()
print("Multiplication:", end=" ")
c1.multiply(c2).display()
```

Q19. Define a class Time having data members hour, minute and second. The class should have a parameterized constructor to initialize its data members. It should also have methods displayTime() to display the time in HH:MM: SS format and addTime() to add two-time objects.

```python
class Time:
    def __init__(self, hour, minute, second):
        self.hour = hour
        self.minute = minute
        self.second = second

    def display_time(self):
        print(f"{self.hour:02}:{self.minute:02}:{self.second:02}")

    def add_time(self, other):
        total_seconds = self.second + other.second
        total_minutes = self.minute + other.minute + total_seconds // 60
        total_hours = self.hour + other.hour + total_minutes // 60
        return Time(total_hours % 24, total_minutes % 60, total_seconds % 60)


t1 = Time(int(input("Enter hour: ")), int(input("Enter minute: ")), int(input("Enter second: ")))
t2 = Time(int(input("Enter hour: ")), int(input("Enter minute: ")), int(input("Enter second: ")))
t1.display_time()
t2.display_time()
t1.add_time(t2).display_time()
```

Q21. Define a class Stack, which has the member function to perform basic operation of stack. Demonstrate the basic operation

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if len(self.items) == 0:
            print("Stack is empty.")
            exit(1)
            
        self.items.pop()

    def peek(self):
        if len(self.items) == 0:
            print("Stack is empty.")
            exit(1)
            
        print("Top:", self.items[-1])

    def display(self):
        print("Stack:", self.items)


stack = Stack()
n = int(input("Enter number of items to push: "))

for i in range(n):
    stack.push(int(input("Enter item: ")))

stack.display()
stack.peek()
n = int(input("Enter number of items to pop: "))

for i in range(n):
    stack.pop()

stack.display()
```

Q22. Define a class Student as follows.  
     Data Mambers
        studName
        studRoll
        studMarks[5]
        studBranch
        studCollegeName
    Member Functions
        dispStudentDetails()
    Create an array of 10 student object. Display the details information about the student who has secured highest average mark.

```python
import random

class Student:
    def __init__(self, studName, studRoll, studMarks, studBranch, studCollegeName):
        self.studName = studName
        self.studRoll = studRoll
        self.studMarks = studMarks
        self.studBranch = studBranch
        self.studCollegeName = studCollegeName

    def average(self):
        return sum(self.studMarks) / len(self.studMarks)

    def dispStudentDetails(self):
        print(f"Student(name={self.studName}, roll={self.studRoll}, marks={self.studMarks}, average={self.average():.2f}, branch={self.studBranch}, college={self.studCollegeName})")


names = ["Alice", "Bob", "Charlie", "Diana", "Eve", "Frank", "Grace", "Harry", "Ivy", "Jack"]
branches = ["CSE", "ECE", "ME", "CE", "EE"]
colleges = ["MIT", "NIT", "IIT", "VIT", "BITS"]

students = []

for i in range(10):
    student = Student(
            names[i],
            i + 1,
            [random.randint(0, 100) for _ in range(5)],
            random.choice(branches),
            random.choice(colleges)
    )
    students.append(student)

for s in students:
    s.dispStudentDetails()

print("\nTop Student:")
top = max(students, key=lambda s: s.average())
top.dispStudentDetails()
```

Q21. Develop a Python program that will deal with employee information of an organization. Define a class Employee, which has the data members and member functions are follows.  
     Data Mambers
         empName
         empNumber
         basicSal
         DA
         HRA
     Member Functions
         calGrossSal ( )
         showEmpDetails( )
     Create an array of objects. Display those employees, whose gross salary is more than 50,000. Note: You are free to add more number of relevant data member and member function.

```python
import random

class Employee:
    def __init__(self, empName, empNumber, basicSal):
        self.empName = empName
        self.empNumber = empNumber
        self.basicSal = basicSal
        self.DA = basicSal * 0.40
        self.HRA = basicSal * 0.20

    def calGrossSal(self):
        return self.basicSal + self.DA + self.HRA

    def showEmpDetails(self):
        print(f"Employee(name={self.empName}, number={self.empNumber}, basicSal={self.basicSal}, DA={round(self.DA)}, HRA={round(self.HRA)}, grossSal={round(self.calGrossSal())})")


names = ["Alice", "Bob", "Charlie", "Diana", "Eve", "Frank", "Grace", "Harry", "Ivy", "Jack"]

employees = []

for i in range(10):
    employee = Employee(
            names[i],
            i + 1,
            random.randint(20000, 60000)
    )
    employees.append(employee)

for e in employees:
    e.showEmpDetails()

print("\nEmployees with gross salary more than 50,000:")
for e in employees:
    if e.calGrossSal() > 50000:
        e.showEmpDetails()
```
