Q1. Create a Class for a Student  
    Write a class Student with attributes:

    • age  
    • marks

```python
class Student:
    def __init__(self, name, age, marks):
        self.name = name
        self.age = age
        self.marks = marks
    
    def display(self):
        print(f"Student(name={self.name}, age={self.age}, marks={self.marks})")

name = input("Enter the name: ")
age = int(input("Enter the age: "))
marks = float(input("Enter marks: "))
student = Student(name, age, marks)
student.display()
```

Q2. Rectangle Area  
    Create a class Rectangle with:  
    • radius  
    • width  
    Add methods:  
    • area() → returns area  
    • perimeter() → returns perimeter

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width
    
    def perimeter(self):
        return 2 * (self.length + self.width)

length = float(input("Enter the length: "))
width = float(input("Enter the width: "))
rectangle = Rectangle(length, width)
print("Area:", rectangle.area())
print("Perimeter:", rectangle.perimeter())
```

Q4. Circle Class  
    Create a class Circle with attribute radius.  
    Add methods:  
    • area()  
    • circumference()  
    Use the value π = 3.14.

```python
class Circle:
    PI = 3.14

    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return Circle.PI * self.radius ** 2
    
    def perimeter(self):
        return 2 * Circle.PI * self.radius
    
radius = float(input("Enter the radius: "))
circle = Circle(radius)
print("Area:", circle.area())
print("Perimeter:", circle.perimeter())
```

Q5. Bank Account System  
    Create a class BankAccount with:  
    • account_holder  
    • balance  
    Methods:  
    • deposit(amount)  
    • withdraw(amount)  
    • display_balance()  
    Add a condition so withdrawal cannot exceed balance. 

```python
class BankAccount:
    def __init__(self, account_holder):
        self.account_holder = account_holder
        self.balance = 0
    
    def deposit(self, amount):
        if amount < 0:
            print("Amount cannot be negetive")
            exit(1)

        self.balance += amount

    def withdraw(self, amount):
        if amount < 0:
            print("Amount cannot be negetive")
            exit(1)
        
        if self.balance < amount:
            print("Insufficient Funds")
            exit(1)
            
        self.balance -= amount
    
    def display_balance(self):
        print("Balance:", self.balance)

account_holder = input("Enter the name of the account holder: ")
bank_account = BankAccount(account_holder)
amount = float(input("Enter the amount to deposit: "))
bank_account.deposit(amount)
amount = float(input("Enter the amount to wthdraw: "))
bank_account.withdraw(amount)
bank_account.display_balance()
```

Q6. Employee Salary  
    Create a class Employee with:  
    • name  
    • salary  
    Add a method increase_salary(percent) that increases salary by given percentage.

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def increase_salary(self, percentage):
        self.salary *= 1 + percentage / 100

    def display(self):
        print(f"Employee(name={self.name}, salary={self.salary})")
    
name = input("Enter name: ")
salary = float(input("Enter salary: "))
employee = Employee(name, salary)
percentage = float(input("Enter the percentage increase in salary: "))
employee.increase_salary(percentage)
employee.display()
```

Q7. Book Class  
    Create a class Book with:  
    • title  
    • author  
    • price  
    Create a method apply_discount(percent) that reduces the price. 

```python
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price
    
    def apply_discount(self, percentage):
        self.price *= 1 - percentage / 100
    
    def display(self):
        print(f"Book(title={self.title}, author={self.author}, price={self.price})")

name = input("Enter the title: ")
author = input("Enter the author: ")
price = float(input("Enter the price: "))
book = Book(name, author, price)
percentage = float(input("Enter the percentage of discount to apply: "))
book.apply_discount(percentage)
book.display()
```

Q8. Temperature Converter  
    Create a class Temperature with attribute celsius.  
    Methods:  
    • to_fahrenheit()  
    • to_kelvin() 

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius
    
    def to_fahrenheit(self):
        return 9 / 5 * self.celsius + 32
    
    def to_kelvin(self):
        return 273.15 + self.celsius

celsius = float(input("Enter temperature in celsius: "))
temperature = Temperature(celsius)
print("Fahrenheit:", temperature.to_fahrenheit())
print("Kelvin:", temperature.to_kelvin())
```

Q9. Library Management  
    Create a class Library with:  
    • book_list  
    Methods:  
    • add_book(book)  
    • remove_book(book)  
    • display_books()

```python
class Library:
    def __init__(self):
        self.books = []
    
    def add_book(self, book):
        self.books.append(book)
    
    def remove_book(self, book):
        if book in self.books:
            self.books.remove(book)

        else:
            print(book, "not present in library")
    
    def display_books(self):
        print("Books: ", self.books)

library = Library()
book = input("Enter a book to add: ")
library.add_book(book)
book = input("Enter a book to remove: ")
library.remove_book(book)
library.display_books()
```

Q10. Shopping Cart  
     Create a class Cart with:  
     • items (dictionary)  
     Methods:  
     • add_item(item, price)  
     • remove_item(item)  
     • total_price() 

```python
class Cart:
    def __init__(self):
        self.items = {}
        
    def add_item(self, item, price):
        self.items[item] = price
    
    def remove_item(self, item):
        if item in self.items:
            self.items.pop(item)
        else:
            print(item, "not present in cart")

    def total_price(self):
        return sum(self.items.values())
    
cart = Cart()
element = eval(input("Enter the item and price to add: "))
cart.add_item(element[0], element[1])
item = input("Enter the item to remove: ")
cart.remove_item(item)
print("Total price:", cart.total_price())
```
