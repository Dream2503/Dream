Q1. Write a Python program to remove duplicates from a list

```python
lst = list(eval(input("Enter a list: ")))
res = []

for element in lst:
    if element not in res:
        res.append(element)

print("The list with unique elements are:", res)
```

Q2. Write a Python program to find the second largest number in a list.

```python
lst = list(eval(input("Enter a list: ")))
max = res = lst[0]

for element in lst[1:]:
    if element > max:
        res = max
        max = element

if res == max:
    print("There is no second maximum element in the list")
else:
    print("The second maximum element in the list is:", res)
```

Q3. Write a Python program to sort a list of elements using the bubble sort algorithm. 

```python
lst = list(eval(input("Enter a list: ")))

for i in range(len(lst) - 1):
    for j in range(len(lst) - i - 1):
        if lst[j] > lst[j + 1]:
            lst[j], lst[j + 1] = lst[j + 1], lst[j]

print("The list after bubble sort is:", lst)
```

Q4. Write a Python program to find the common elements between two lists.

```python
lst1 = list(eval(input("Enter a list1: ")))
lst2 = list(eval(input("Enter a list2: ")))
res = [element for element in lst1 if element in lst2]
print("The intersection of the two list are:", res)
```

Q5. Write a Python program to count the frequency of each element in a list.

```python
lst = list(eval(input("Enter a list: ")))
seen = []
print("The frequency of all the element in the list is:", end=" ")

for element in lst:
    if element not in seen:
        print(element, "->", lst.count(element), sep="", end=" ")
        seen.append(element)
```

Q6. Write a Python program to count the number of strings where the string length is 2 or more and the first and last character are same from a given list of strings

```python
lst = list(eval(input("Enter a list: ")))
res = [element for element in lst if len(element) >= 2 and element[0] == element[-1]]
print("The resultant list is:", res)
```

Q7. Write a Program that get two lists as input and check if they have at least one common member

Q8. Write a Python program to print a specified list after removing the 0th, 4th and 5th elements. 

```python
lst = list(eval(input("Enter a list: ")))

if len(lst) > 5:
    for idx in (5, 4, 0):
        lst.pop(idx)
    
    print("The resultant list is:", lst)
else:
    print("There must be atleast 6 element in the list")
```

Q10.  Write a Python Program to print all Possible Combinations from the three Digits

```python
lst = list(eval(input("Enter the 3 digits: ")))
print("The the combination of the 3 digits,", lst, "are:")

for first in lst:
    for middle in lst:
        for last in lst:
            print(first, middle, last, sep="", end=" ")
        print()
```
