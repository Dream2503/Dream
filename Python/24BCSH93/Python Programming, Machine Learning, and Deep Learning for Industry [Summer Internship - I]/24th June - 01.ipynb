# Checking for outliers using IQR method

```python
df.describe().T
```

```python
q1 = df.describe().T.loc[:, "25%"]
q1
```

# extracting quartile 3

```python
q3 = df.describe().T.loc[:, "75%"]
q3
```

```python
IQR = q3 - q1
IQR
```

```python
lower_boundary = np.round(q1 - 1.5 * IQR, 2)
upper_boundary = np.round(q3 + 1.5 * IQR, 2)
```

```python
lower_boundary
```

```python
upper_boundary
```

```python
cols = df.select_dtypes(exclude='object')
```

# finding the quartile directly

```python
Q1 = df_cols.quantile(0.25)
Q3 = df_cols.quantile(0.75)
```

```python
Q1
```

```python
count = 0

for i in cols:
    temp = (cols[i] > upper_boundary[i]) | (cols[i] < lower_boundary[i])
    print(f'outliers in [i] are :\n {temp.value_counts()}')
    print("-----------------------------------------------")
    count += 1
```
