# Split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train = train_df.drop(columns=[187])
y_train = train_df[187]
X_test = test_df.drop(columns=[187])
y_test = test_df[187]

X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

# normalize

```python
scaler = StandardScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_valid_normalized = scaler.transform(X_valid)
X_test_normalized = scaler.transform(X_test)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

clf_dt = DecisionTreeClassifier(random_state=42)
clf_dt.fit(X_train_normalized, y_train)
```

```python
y_pred_valid = clf_dt.predict(X_valid_normalized)
accuracy_valid = accuracy_score(y_valid, y_pred_valid)
accuracy_valid
```

```python
y_pred_test = clf_dt.predict(X_test_normalized)
accuracy_test = accuracy_score(y_test, y_pred_test)
accuracy_test
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

clf_rf = RandomForestClassifier(50, random_state=123)
clf_rf.fit(X_train_normalized, y_train)
```

```python
y_pred_valid = clf_rf.predict(X_valid_normalized)
accuracy_valid = accuracy_score(y_valid, y_pred_valid)
accuracy_valid
```

```python
y_pred_test = clf_rf.predict(X_test_normalized)
accuracy_test = accuracy_score(y_test, y_pred_test)
accuracy_test
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

clf_nb = GaussianNB()
clf_nb.fit(X_train_normalized, y_train)
```

```python
y_pred_valid = clf_nb.predict(X_valid_normalized)
accuracy_valid = accuracy_score(y_valid, y_pred_valid)
accuracy_valid
```

```python
y_pred_test = clf_nb.predict(X_test_normalized)
accuracy_test = accuracy_score(y_test, y_pred_test)
accuracy_test
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

clf_knn = KNeighborsClassifier()
clf_knn.fit(X_train_normalized, y_train)
```

```python
y_pred_valid = clf_knn.predict(X_valid_normalized)
accuracy_valid = accuracy_score(y_valid, y_pred_valid)
accuracy_valid
```

```python
y_pred_test = clf_knn.predict(X_test_normalized)
accuracy_test = accuracy_score(y_test, y_pred_test)
accuracy_test
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

print("Confusion Matrices:")
print("\n1. Decision Tree:\n", confusion_matrix(y_test, clf_dt.predict(X_test_normalized)))
print("\n2. Random Forest:\n", confusion_matrix(y_test, clf_rf.predict(X_test_normalized)))
print("\n3. Naive Bayes:\n", confusion_matrix(y_test, clf_nb.predict(X_test_normalized)))
print("\n4. KNN:\n", confusion_matrix(y_test, clf_knn.predict(X_test_normalized)))
```
