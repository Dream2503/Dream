```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mu, sigma = 0, 1
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, mu, sigma)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Normal Distribution")
plt.fill_between(x, y, where=(x >= -1) & (x <= 1), alpha=0.3, label="68% within one sigma")
plt.fill_between(x, y, where=(x >= -2) & (x <= 2), alpha=0.2, label="95% within two sigma")
plt.fill_between(x, y, where=(x >= -3) & (x <= 3), alpha=0.1, label="99.7% within two sigma")
plt.title("Standard Normal Distribution")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.stats import skew

np.random.seed(0)
normal_data = np.random.normal(0, 1, 1000)
pos_skewed = np.random.exponential(1, 1000)
neg_skewed = -np.random.exponential(1, 1000)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, data, title in zip(axes, [normal_data, pos_skewed, neg_skewed], ["Normal (Skew = 0)", "Positive Skew", "Negative Skew"]):
    ax.hist(data, bins=30, color="skyblue", edgecolor="black")
    ax.set_title(f"{title}\nSkewness: {skew(data):.2f}")

plt.tight_layout()
plt.show()
```
