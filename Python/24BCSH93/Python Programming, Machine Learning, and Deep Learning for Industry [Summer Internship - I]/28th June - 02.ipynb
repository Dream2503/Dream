### Import important libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Generate Datapoints:

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0], [1], [1], [0]])
```

```python
X
```

```python
Y
```

### Building NN Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
```

```python
model.add(Dense(4, "relu"))
```

```python
model.add(Dense(1, "sigmoid"))
```

#### Compilation: Preparing the NN for training: how the learning is happening, what to minimise and which performance metric to track

```python
model.compile(Adam(0.1), "mse", metrics=["accuracy"])
```

#### Training the model

```python
model.fit(X, Y, epochs=500, batch_size=8192)
```

#### Testing the model

```python
predictions = model.predict(X)
```

```python
predictions
```

```python
predictions = np.round(predictions)
```
