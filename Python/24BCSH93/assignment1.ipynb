Question 16

```python
a = float(input("Enter the coefficient of x^2: "))
b = float(input("Enter the coefficient of x: "))
c = float(input("Enter the constant coefficient: "))

determinant = (b ** 2 - 4 * a * c) * 0.5
print("First Root:", (-b + determinant) / (2 * a))
print("Second Root:", (-b - determinant) / (2 * a))
```

Question 20

```python
n1 = int(input("Enter the first number: "))
n2 = int(input("Enter the second number: "))

print("Logical NOT:", not n1)
print("Logical AND:", n1 and n2)
print("Logical OR:", n1 or n2)
print("Logical XOR:", (n1 and not n2) or (not n1 and n2))
print("Logical XNOR:", (n1 and n2) or (not n1 and not n2))
print()
print("Bitwise NOT: ", ~n1)
print("Bitwise AND:", n1 & n2)
print("Bitwise OR:", n1 | n2)
print("Bitwise XOR:", n1 ^ n2)
print("Bitwise XNOR:", ~(n1 ^ n2))
print("Bitwise Left Shift:", n1 << n2)
print("Bitwise Right Shift:", n1 >> n2)
```
