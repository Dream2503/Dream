```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = 2
c = 3
np.random.seed(42)
X = np.random.rand(500)
y = m * X + c + np.random.randn(500)
```

```python
plt.plot(X, y, 'bo')
plt.show()
```

```python
split_ratio = 0.8
split_index = int(split_ratio * len(X))
```

```python
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]
```

```python
weights = np.random.randn(2)
```

```python
learning_rate = 0.01
epochs = 1000
```

```python
for epoch in range(epochs):
    y_pred = weights[0] + weights[1] * X_train
    error = y_pred - y_train
    gradient_intercept = np.mean(error)
    gradient_slope = np.mean(error * X_train)
    weights[0] -= learning_rate * gradient_intercept
    weights[1] -= learning_rate * gradient_slope
```

```python
y_pred_test = weights[0] + weights[1] * X_test
mse = np.mean((y_pred_test - y_test) ** 2)
```

```python
# Plot the data points and linear regression line
plt.scatter(X, y, label="Data Points")
plt.plot(X_test, y_pred_test, color="red", label="Linear Regression Line")

# Set plot labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Show the plot
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = np.random.rand(100, 1) * 10
y = 2 * X + 3 + np.random.randn(100, 1)
```

```python
# Creating the scatter plot
plt.scatter(X, y, label="Data points")

# Adding labels and title
plt.xlabel('X')
plt.ylabel('y')
plt.title("Scatter Plot of X and y")

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

```python
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)


def linear_regression(input_size, output_size):
    return nn.Linear(input_size, output_size)


input_size = 1
output_size = 1
model = linear_regression(input_size, output_size)
```

```python
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
num_epochs = 1000

for epoch in range(num_epochs):
    y_pred = model(X_tensor)
    loss = criterion(y_pred, y_tensor)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 100 == 0:
        print(f"Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}")
```

```python
# Parameters
num_points = 200
x_range = np.linspace(0, 10, num_points)
noise = np.random.normal(0, 0.3, num_points)

# Generate zigzag points
y = np.sin(x_range) + noise

plt.scatter(x_range, y, label="Zigzag Points")
plt.title("Zigzag Data Points")
plt.legend()
plt.show()
```

```python
# Define a simple neural network using a sequential module
model = nn.Sequential(
        nn.Linear(1, 50),  # Input layer: 1 input feature, 50 neurons
        nn.Tanh(),  # Activation function
        nn.Linear(50, 1)  # Output layer: 50 neurons, 1 output feature
)
```

```python
# Training loop
num_epochs = 10000
x_data = torch.tensor(x_range, dtype=torch.float32)
y_data = torch.tensor(y, dtype=torch.float32)

for epoch in range(num_epochs):
    # Forword pass
    # print(x_data.unsqueeze(1).shape)
    y_pred = model(x_data.unsqueeze(1))
    loss = criterion(y_pred, y_data.unsqueeze(1))

    # Backpropagation and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 100 == 0:
        print(f"Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}")
```
